### Importing libraries:

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib.inline
import numpy as np
```

### Indian crops dataset:

```python
df = pd.read_csv('Indian crops dataset.csv')
df.head(10)
```

### Dataset shortlisted:

```python
df1 = pd.read_excel('Indian crops shortlisted.xlsx')
df1.head(10)
```

### Correlation between Year & Rice area, Rice production, Rice yield, Rice share:

```python
df1[['Year','rice_area_1000ha','rice_production_1000tons','rice_yield_kgperha','rice_share_total_percent']].corr()
```

### Correlation between Wheat & Wheat area, Wheat production, Wheat yield, Wheat share:

```python
df1[['Year','wheat_area_1000ha','wheat_production_1000tons','wheat_yield_kgperha','wheat_share_total_percent']].corr()
```

### Correlation between Year & Sugarcane area, production, yield, share:

```python
df1[['Year','sugarcane_area_1000ha','sugarcane_production_1000tons','sugarcane_yield_kgperha','sugarcane_share_total_percent']].corr()
```

### Correlation between Year and Fruits area, share, Vegetables area, share, Potatoes area, Onion area:

```python
df1[['Year','fruits_area_1000ha','fruits_share_total_percent','vegetables_area_1000ha','vegetables_share_total_percent','fruits_vegetables_area_1000ha','potatoes_area_1000ha','onion_area_1000ha']].corr()
```
